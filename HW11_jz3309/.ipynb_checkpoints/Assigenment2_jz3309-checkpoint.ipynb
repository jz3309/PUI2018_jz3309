```python
collision = pd.read_csv("%s/NYPDMotorVehicleCollisions.csv"%os.getenv("PUIDATA"))
```

```python
collision1 = collision[['DATE','ZIP CODE','TIME','BOROUGH','LATITUDE','LONGITUDE','LOCATION','UNIQUE KEY']]
```

```python
collision1.head()
```

```python
collision1.dropna(axis=0, inplace=True)
collision1['DATE'] = pd.to_datetime(collision1['DATE'])
a = collision1.groupby('ZIP CODE').count()
```

```python
collision1.size
```

```python
b = a.reset_index()
b.rename(columns={'DATE':'amount'},inplace=True)
b.head()
```

```python
mapdata = b[['ZIP CODE','amount']].merge(collision1[['ZIP CODE','LATITUDE','LONGITUDE','LOCATION']],on='ZIP CODE', how='left')

mapdata.head()
```

```python
b.dtypes
```

```python
amount = b[['ZIP CODE','amount']]
amount.head()
```

```python
zips = gpd.GeoDataFrame.from_file(os.getenv("PUIDATA") + "/" + "ZIP_CODE_040114.shp")
zips.head()
```

```python
zips.dtypes
zips['ZIPCODE']=zips['ZIPCODE'].astype(int)
```

```python
plotdata = zips.merge(amount,left_on='ZIPCODE',right_on='ZIP CODE')
plotdata.head()
```

```python
plotdata.columns
```

## Figure plotted in HW8

```python
fig, ax = plt.subplots(figsize=(13,10))
ax.axis('off')
plotdata.plot(column='amount',legend=True, cmap='Oranges',ax=ax)
```

## Figure1:The distribution of motor vehicle collisions from 2012 to 2018 by zipcode
## The number of vehicle collisions happened in Lower Manhattan and Brooklyn downtown are larger than other areas.


## Improve the plot

```python
fig= plt.figure(figsize=(25,10))

ax1 = fig.add_subplot(121)
ax1.axis('off')
plotdata.plot(column=u'amount',legend=True,ax=ax1)
ax1.set_title('Numeber of collisions by zipcode',fontsize=20)

ax2 = fig.add_subplot(122)
ax2.axis('off')
plotdata.plot(column=u'POPULATION',legend=True,ax=ax2)
ax2.set_title('Population by zipcode',fontsize=20)
```

First figure shows the number of collisions by zipcode. Darker areas means more collisions happened in the two years.Northeast of Brooklyn has largst number of collisions. Second figure shows the populaiton by zipcode, I try to investigate the population density in the areas as there may be more accidents.in brooklyn, the ares has more collisions truly has larger population.
